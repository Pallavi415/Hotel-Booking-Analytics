# Multivariate Analysis

## Objective

In this notebook, I analyze the interaction between multiple features to uncover deeper patterns in customer behavior, booking trends, and pricing. These insights help identify relationships that cannot be observed through univariate or bivariate analysis alone.

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

hotel_df = pd.read_csv("../data/hotel_bookings_cleaned.csv")
```

Analysis 1: Correlation Analysis

### Why am I analyzing this?
A correlation matrix helps identify the strength of relationships between numerical features. It provides an overview of how variables are associated with one another and highlights the features that may influence booking behavior.

```python
numerical_features = hotel_df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(12, 8))

sns.heatmap(
    numerical_features.corr(),
    cmap="coolwarm",
    annot=True,
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Heatmap")
plt.tight_layout()

plt.savefig(
    "../visuals/correlation_heatmap.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
plt.show()
```

### Observation

Most numerical features have weak to moderate correlations, while only a few feature pairs show stronger relationships.

### Insight

The heatmap provides a quick overview of feature relationships and helps identify variables that may be useful for predictive modeling.

Analysis 2: Cancellation Rate by Hotel and Deposit Type

### Why am I analyzing this?
Cancellation behavior may vary across hotel types and deposit policies. Analyzing these features together helps identify combinations associated with higher cancellation rates.

```python
plt.figure(figsize=(10, 6))

sns.countplot(
    data=hotel_df,
    x="hotel",
    hue="deposit_type"
)

plt.title("Hotel Type and Deposit Policy")
plt.xlabel("Hotel Type")
plt.ylabel("Number of Bookings")
plt.legend(title="Deposit Type")

plt.tight_layout()
plt.show()
```

### Observation

The distribution of deposit policies differs across hotel types, indicating variations in booking practices.

### Insight

Understanding the relationship between hotel type and deposit policy can support better booking management and cancellation control.

Analysis 3: Room Pricing Across Customer Types

### Why am I analyzing this?
Room prices may vary depending on both the hotel type and the type of customer. Analyzing these features together helps identify customer groups associated with higher room rates.

```python
plt.figure(figsize=(12, 6))

sns.boxplot(
    data=hotel_df,
    x="customer_type",
    y="adr",
    hue="hotel"
)

plt.title("ADR by Customer Type and Hotel")
plt.xlabel("Customer Type")
plt.ylabel("Average Daily Rate (ADR)")
plt.legend(title="Hotel")

plt.tight_layout()
plt.savefig("../visuals/adr_customer_type_hotel.png", dpi=150)

plt.show()
```

### Observation

The distribution of ADR varies across customer types, and the pricing pattern differs between city and resort hotels.

### Insight

Comparing pricing across customer segments helps identify high-value customers and supports more effective revenue management strategies.

Analysis 4: Lead Time by Market Segment and Booking Status

### Why am I analyzing this?
Lead time can vary across market segments and may influence booking cancellations. Examining these features together provides a better understanding of customer booking behavior.

```python
plt.figure(figsize=(14, 6))

sns.boxplot(
    data=hotel_df,
    x="market_segment",
    y="lead_time",
    hue="is_canceled"
)

plt.title("Lead Time by Market Segment")
plt.xlabel("Market Segment")
plt.ylabel("Lead Time (Days)")
plt.xticks(rotation=45)
plt.legend(title="Booking Status", labels=["Not Cancelled", "Cancelled"])

plt.tight_layout()
plt.savefig(
    "../visuals/lead_time_market_segment.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

### Observation

Lead time varies across market segments, and cancelled bookings generally show greater variation than completed bookings.

### Insight

Analyzing lead time across different customer segments helps identify booking patterns that may contribute to cancellation risk.

Analysis 5: Special Requests Across Hotel and Customer Types

### Why am I analyzing this?
Special requests reflect customer preferences and service expectations. Analyzing them across hotel and customer types helps understand how guest requirements differ among customer groups.

```python
plt.figure(figsize=(12, 6))

sns.boxplot(
    data=hotel_df,
    x="customer_type",
    y="total_of_special_requests",
    hue="hotel"
)

plt.title("Special Requests by Customer Type and Hotel")
plt.xlabel("Customer Type")
plt.ylabel("Number of Special Requests")
plt.legend(title="Hotel")

plt.tight_layout()
plt.savefig("../visuals/special_requests_customer_type.png", dpi=150)
plt.show()
```

### Observation

The number of special requests varies across customer types, with noticeable differences between city and resort hotels.

### Insight

Understanding guest requirements across customer segments helps hotels improve service planning and deliver a better customer experience.

These analyses provide a comprehensive understanding of customer behavior and support data-driven business decisions.
